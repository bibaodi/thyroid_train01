```python
def statis_type_inModelOut(csvfile:pathlib.Path, className:str, scorethreshold:float):
    if not csvfile.exists():
        print(f"not exist: {csvfile}")
        return
    modelO_pd = pd.read_csv(csvfile)
    print(f"{csvfile}, ORIGIN:", modelO_pd.index)
    #for iimgidx 
    isInClass = modelO_pd['class'].str.startswith(className)
    isScoreGood =  modelO_pd['probability']>scorethreshold
    filtered_df = modelO_pd[isInClass & isScoreGood]
    
    print(f"{className}_filterd={filtered_df}")

for icsv in modelOutCsvs:
    mocsv = pathlib.Path(inputDataFolder) / pathlib.Path(icsv)
    statis_type_inModelOut(mocsv, 'isthmus', 0.99)
```

```python
def statis_types_inModelOut(csvfile:pathlib.Path, className:str, scorethreshold:float):
    if not csvfile.exists():
        print(f"not exist: {csvfile}")
        return
    modelO_pd = pd.read_csv(csvfile)
    
    classSeriesAll = modelO_pd['class']
    isScoreGood =  modelO_pd['probability']>scorethreshold
    filtered_df = modelO_pd[isScoreGood]
    classSeriesFilted = filtered_df['class']
    
    print(f"{csvfile}, ORIGIN:", classSeriesAll.value_counts())
    print(f"classSeriesFilted:", classSeriesFilted.value_counts())

for icsv in modelOutCsvs:
    mocsv = pathlib.Path(inputDataFolder) / pathlib.Path(icsv)
    statis_types_inModelOut(mocsv, 'isthmus', 0.99)
```

```python
def filter_by_position_prefix(df, prefix: str) -> pd.DataFrame:
    """Filter dataframe where position column starts with given prefix (case-insensitive)"""
    prefix = prefix.lower()
    return df[df['position'].str.startswith(prefix)].copy()

filtered_df = filter_by_position_prefix(df, r'峡部')

```

```python
def matchCsvAndExcel(modelO, df):
    imgname_pos_match_isthmus_list=[];
    isthmuxCnt=0
    isthmusMatch=0
    for iimgidx in modelO.index:
        sop_uid = modelO.at[iimgidx, 'image_path']
        imgsop_uid = sop_uid[0:len('02.202201220336.01.20188.0003')]
    
        clsName = modelO.at[iimgidx, 'class']
        if 'isthmus' == clsName:
            isthmuxCnt+=1
            #print(imgsop_uid, clsName)
    
            for jidx in df.index:
                j_sopuid = df.at[jidx, 'sop_uid']
                j_pos = df.at[jidx, 'position']
                if j_sopuid.startswith(imgsop_uid) and j_pos.startswith(r'峡部'):
                    isthmusMatch +=1
                    imgname_pos_match_isthmus_list.append((sop_uid, clsName, j_sopuid, j_pos))
                    print(f"found: {j_sopuid}, {j_pos}")
            if iimgidx>9 and isthmuxCnt>9 and isthmusMatch>9999:
                break
    return imgname_pos_match_isthmus_list

imgname_pos_match_isthmus_list = matchCsvAndExcel(modelO, filtered_df)
```

```python
output_file=pathlib.Path(inputDataFolder) /"foundAllIsthmus-single2.csv"

def writeListToCsvFile(output_file, imgname_pos_match_isthmus_list):
    with open(output_file, 'w') as f:
        f.write("ImageName, predCls, sopuid, DataLabel\n")
        for irow in imgname_pos_match_isthmus_list:
            csvRow = ','.join(irow)
            f.write(f"{csvRow}\n")

writeListToCsvFile(output_file, imgname_pos_match_isthmus_list)
```
